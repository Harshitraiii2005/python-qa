# Python Q&A Assistant — Test Results

Evaluating 10 diverse queries against the live API.


```python
import httpx
import json
from IPython.display import Markdown, display

BASE = 'http://localhost:8000'

queries = [
    'How do I reverse a list in Python?',
    'How do I read a CSV file with pandas?',
    'What is the difference between __str__ and __repr__?',
    'How do I merge two dictionaries in Python 3?',
    'How do I use @property decorator in Python?',
    'What is the difference between asyncio.gather and asyncio.wait?',
    'How do I profile Python code to find bottlenecks?',
    'What is Python?',
    'How do I unpack nested list comprehensions with walrus operator?',
    'How do I install Node.js packages?',
]

results = []
for q in queries:
    r = httpx.post(f'{BASE}/ask', json={'question': q}, timeout=30)
    data = r.json()
    results.append(data)
    display(Markdown(f'### Q: {q}'))
    display(Markdown(f'**Answer:** {data["answer"][:500]}...'))
    display(Markdown(f'**Top source:** {data["sources"][0]["title"]} (relevance: {data["sources"][0]["relevance"]})'))
    display(Markdown(f'**Latency:** {data["latency_ms"]}ms'))
    print('---')
```

## Summary Table


```python
import pandas as pd
summary = pd.DataFrame([{
    'Question': r['question'][:60],
    'Answer length': len(r['answer']),
    'Sources': len(r['sources']),
    'Top relevance': r['sources'][0]['relevance'] if r['sources'] else 0,
    'Latency (ms)': r['latency_ms'],
} for r in results])
display(summary.to_markdown(index=False))
```
